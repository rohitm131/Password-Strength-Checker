```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('data.csv',',',error_bad_lines=False)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df[df['strength'].isnull()]
```

```python
df.dropna(inplace=True)
```

```python
passwords_tuple=np.array(df)
```

```python
passwords_tuple
```

```python
import random
random.shuffle(passwords_tuple) #shuffling data for robustness
```

```python
X = [labels[0] for labels in passwords_tuple]
y = [labels[1] for labels in passwords_tuple]
```

```python
X,y
```

```python
import seaborn as sns
```

```python
sns.set_style('whitegrid')
sns.countplot(x='strength',data=df,palette='RdBu_r')
```

```python
df.shape
```

```python
def word_divide_char(inputs):
    characters=[]
    for i in inputs:
        characters.append(i)
    return characters
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer=TfidfVectorizer(tokenizer=word_divide_char)
X=vectorizer.fit_transform(X)
```

```python
X.shape
```

```python
vectorizer.vocabulary_
```

```python
##Logistic Regression

from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
X_train,X_tet,y_train,y_test=train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
lr=LogisticRegression(penalty='l2',multi_class='ovr')
lr.fit(X_train,y_train)
```

```python
print(lr.score(X_tet,y_test))
```

```python
#Multinomial

clf=LogisticRegression(random_state=0, multi_class='multinomial', solver='newton-cg')
clf.fit(X_train,y_train)
```

```python
print(clf.score(X_tet,y_test))
```

```python
X_predict=np.array(["%@123abcd"])
X_predict=vectorizer.transform(X_predict)
y_pred=lr.predict(X_predict)
print(y_pred)
```

```python
from sklearn.naive_bayes import MultinomialNB
nb=MultinomialNB()
nb.fit(X_train,y_train)
```

```python
nb.score(X_tet,y_test)
```
